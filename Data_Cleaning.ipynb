### Code to Delete a Class Instance from a `.txt` File

```python
label_dir = "temp"
count_0 = 0
count_1 = 0
count_2 = 0
count_3 = 0
count_4 = 0  # This should end up as 0 after deleting rows with class ID "4"
count_5 = 0

for fname in os.listdir(label_dir):
    if fname.endswith(".txt"):
        try:
            # Open the file to read all rows
            with open(os.path.join(label_dir, fname), "r") as f:
                reader = csv.reader(f, delimiter=" ")
                rows = [row for row in reader if row]  # Ensure each row is not empty

            # Filter out rows with class ID "4"
            filtered_rows = [row for row in rows if row[0] != "4"]

            # Write back only the filtered rows to the file
            with open(os.path.join(label_dir, fname), "w") as f:
                writer = csv.writer(f, delimiter=" ")
                writer.writerows(filtered_rows)

            # Count occurrences of each class in the updated file
            for row in filtered_rows:
                class_id = row[0]
                
                if class_id == "0":
                    count_0 += 1
                # elif class_id == "1":
                #     count_1 += 1
                elif class_id == "2":
                    count_2 += 1
                elif class_id == "3":
                    count_3 += 1
                # elif class_id == "5":
                #     count_5 += 1

        except Exception as e:
            print(f"Error processing file {fname}: {e}")
            continue

# Print final counts after deletion
print(f"""
Yawn: {count_0}
Drowsy: {count_1}
EyesClosed: {count_2}
HeavyEyes: {count_3}
EyesOpened: {count_4}  # Should be 0 after deletion
NotDrowsy: {count_5}
""")
```

### Code to Check for Missing Pairs

```python
import os

# Specify the folder path
folder_path = 'Driver Drowsiness Complete Dataset'

# Get lists of all .txt and .png files (without extensions) in the folder
txt_files = {os.path.splitext(f)[0] for f in os.listdir(folder_path) if f.endswith('.txt')}
png_files = {os.path.splitext(f)[0] for f in os.listdir(folder_path) if f.endswith('.png')}

# Find missing pairs
missing_txt_files = png_files - txt_files
missing_png_files = txt_files - png_files

# Delete files missing their pairs and print confirmation
for name in missing_txt_files:
    png_file = os.path.join(folder_path, f"{name}.png")
    try:
        os.remove(png_file)
        print(f"Deleted missing pair file: {png_file}")
    except Exception as e:
        print(f"Could not delete {png_file}. Error: {e}")

for name in missing_png_files:
    txt_file = os.path.join(folder_path, f"{name}.txt")
    try:
        os.remove(txt_file)
        print(f"Deleted missing pair file: {txt_file}")
    except Exception as e:
        print(f"Could not delete {txt_file}. Error: {e}")

if not missing_txt_files and not missing_png_files:
    print("No files were missing pairs. All pairs are present.")
else:
    print("Deletion of unpaired files complete.")
```

### Code to Just Count Class Instance

```python
label_dir = "Driver Drowsiness Complete Dataset"
count_0 = 0
count_1 = 0
count_2 = 0
count_3 = 0
count_4 = 0
count_5 = 0

for fname in os.listdir(label_dir):
    if fname.endswith(".txt"):
        try:
            with open(os.path.join(label_dir, fname), "r") as f:
                reader = csv.reader(f, delimiter=" ")

                # Loop through each row and count class instances
                for row in reader:
                    # Ensure the row has at least one element to avoid index errors
                    if len(row) > 0:
                        if row[0] == "0":
                            count_0 += 1
                        elif row[0] == "1":
                            count_1 += 1
                        elif row[0] == "2":
                            count_2 += 1
                        elif row[0] == "3":
                            count_3 += 1
                        elif row[0] == "4":
                            count_4 += 1
                        elif row[0] == "5":
                            
                            count_5 += 1
                            

        except Exception as e:
            print(f"Error processing file {fname}: {e}")
            continue

# Print counts
print(f"""
Yawn: {count_0}
Drowsy: {count_1}
EyesClosed: {count_2}
HeavyEyes: {count_3}
EyesOpened: {count_4}
NotDrowsy: {count_5}
""")
```

### Code to Move `.txt` Files with Yawn (0), Eyesclosed (2) or HeavyEyes (3) IDs to New Folder `affected`

```python
label_dir = "Yawn"
affected_dir = os.path.join(label_dir, "affected")
os.makedirs(affected_dir, exist_ok=True)  # Create the 'affected' folder if it doesn't exist

# Initialize class counts
count_0 = 0
count_1 = 0
count_2 = 0
count_3 = 0
count_4 = 0
count_5 = 0

for fname in os.listdir(label_dir):
    if fname.endswith(".txt"):
        txt_file_path = os.path.join(label_dir, fname)
        base_name = os.path.splitext(fname)[0]  # Get the base name of the file without extension
        move_to_affected = False  # Flag to decide if the file should be moved

        try:
            with open(txt_file_path, "r") as f:
                reader = csv.reader(f, delimiter=" ")
                
                for row in reader:
                    # Check if row is not empty
                    if len(row) > 0:
                        class_id = row[0]

                        # Count instances of each class ID
                        if class_id == "0":
                            count_0 += 1
                            move_to_affected = True
                        elif class_id == "1":
                            count_1 += 1
                        elif class_id == "2":
                            count_2 += 1
                            move_to_affected = True
                        elif class_id == "3":
                            count_3 += 1
                            move_to_affected = True
                        elif class_id == "4":
                            count_4 += 1
                        elif class_id == "5":
                            count_5 += 1

            # Move the .txt file and corresponding .jpg or .png file if flagged
            if move_to_affected:
                # Move .txt file
                shutil.move(txt_file_path, os.path.join(affected_dir, fname))

                # Check and move the corresponding image file (.jpg or .png)
                for ext in [".jpg", ".png"]:
                    img_file_path = os.path.join(label_dir, base_name + ext)
                    if os.path.exists(img_file_path):
                        shutil.move(img_file_path, os.path.join(affected_dir, base_name + ext))
                        break  # Exit loop once image is moved

        except Exception as e:
            print(f"Error processing file {fname}: {e}")
            continue

# Print counts
print(f"""
Yawn: {count_0}
Drowsy: {count_1}
EyesClosed: {count_2}
HeavyEyes: {count_3}
EyesOpened: {count_4}
NotDrowsy: {count_5}
""")
```
