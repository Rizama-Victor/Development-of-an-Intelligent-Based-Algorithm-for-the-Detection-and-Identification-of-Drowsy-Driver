```python
# Imports the Image Module from the Pillow Library
from PIL import Image
# PIL = Python image library
import os
```

```python
# Function to convert a JPEG file to PNG format
def convert_jpeg_to_png(input_jpeg_image_path, output_png_image_path):
    with Image.open(input_jpeg_image_path) as img:
        img.save(output_png_image_path, format="PNG")
    print(f"Converted {input_jpeg_image_path} to {output_png_image_path}")

# Folder containing the JPEG files to convert
input_folder = "All_Others - Copy"

# Loop through all files in the "affected" folder
for filename in os.listdir(input_folder):
    if filename.endswith(".jpg"):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(input_folder, f"{os.path.splitext(filename)[0]}.png")
        
        # Convert JPEG to PNG
        convert_jpeg_to_png(input_path, output_path)
```
